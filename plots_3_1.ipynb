```
import os
import numpy as np
import glob
import matplotlib.pyplot as plt

dir = '/content/drive/My Drive/DL-assignment-2'
os.chdir(dir)
%matplotlib inline
```

```
def plot_curve(folder_name,label,color):
    epochs = [x for x in range(1,31)]
    # load data
    lc_path = os.path.join(folder_name, "learning_curves.npy")
    x = np.load(lc_path, allow_pickle=True)[()]
    # plot ppl
    ax.plot(epochs, x['train_ppls'], color=color, label=label, linewidth=2)
    ax.plot(epochs, x['val_ppls'], '--', color=color, linewidth=2)

def plot_time_curve(folder_name, label, color):
    # load data
    lc_path = os.path.join(folder_name, "learning_curves.npy")
    x = np.load(lc_path, allow_pickle=True)[()]
    times_ = np.array(x['times'])
    times = times_.copy()
    times = [np.sum(times_[0:i]) for i in range(len(times_))]
    ax.plot(times, x['train_ppls'], color=color, label=label, linewidth=2)
    ax.plot(times, x['val_ppls'], '--', color=color, linewidth=2)
```

```
exp_folders = glob.glob(dir +"/*RNN*")
plt.rcParams['figure.figsize'] = [10, 7]
fig, ax = plt.subplots()
plot_curve(exp_folders[0],label='lr=1.0, bs=128, opt=SGD', color='r')
plot_curve(exp_folders[1],label='lr=1.0, bs=20, opt=SGD', color='b')
plot_curve(exp_folders[2],label='lr=10.0, bs=128, opt=SGD', color='c')
plot_curve(exp_folders[3],label='lr=0.001, bs=128, opt=ADAM', color='m')
plot_curve(exp_folders[4],label='lr=0.0001, bs=128, opt=ADAM', color='g')
ax.plot([x for x in range(1,31)], [10000 for x in range(1,31)], '--', label='validation', color='k', linewidth=2)
ax.set_ylim([0,700])
ax.legend(loc='upper right')
plt.xlabel('Epochs')
plt.ylabel('PPL')
plt.show();

exp_folders = glob.glob(dir +"/*RNN*")
plt.rcParams['figure.figsize'] = [10, 7]
fig, ax = plt.subplots()
plot_time_curve(exp_folders[0],label='lr=1.0, bs=128, opt=SGD', color='r')
plot_time_curve(exp_folders[1],label='lr=1.0, bs=20, opt=SGD', color='b')
plot_time_curve(exp_folders[2],label='lr=10.0, bs=128, opt=SGD', color='c')
plot_time_curve(exp_folders[3],label='lr=0.001, bs=128, opt=ADAM', color='m')
plot_time_curve(exp_folders[4],label='lr=0.0001, bs=128, opt=ADAM', color='g')
ax.plot([x for x in range(1,31)], [10000 for x in range(1,31)], '--', label='validation', color='k', linewidth=2)
ax.set_ylim([0,700])
ax.legend(loc='upper right')
plt.xlabel('Wall-Clock Time')
plt.ylabel('PPL')
plt.show();
```
